```python
import numpy as np
from numpy import exp,arange
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from PIL import Image
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
```

```python
A = np.array([[175,180,175,160,150,130,110,105,100],
              [180,200,180,175,160,150,130,110,105],
              [175,180,175,160,150,130,110,105,100],
              [160,175,160,150,130,110,105,110,105],
              [150,160,150,130,100,105,110,115,110],
              [130,150,130,110,105,110,115,130,115],
              [110,130,110,105,110,115,130,140,130],
              [105,110,105,110,115,130,140,160,140],
              [100,105,100,105,110,115,130,140,130]])

plt.figure(figsize=(5,5))              # tama√±o de imagen
plt.imshow(A, cmap = 'inferno')        # grafico y paleta de colores
im_ratio = A.shape[0]/A.shape[1]       # ancho y alto de la imagen
plt.colorbar(fraction=0.046*im_ratio)  # barra de colores
plt.show()
```

```python
# Funci√≥n a graficar
def z_func(x,y):
    return np.sin(x) + np.sin(y) + 0.01*pow((abs(x)+abs(y)), 2)
 
x = arange(-15.0, 15.0, 0.01)
y = arange(-15.0, 15.0, 0.01)

X,Y = meshgrid(x, y)  # grid o matriz de puntos x,y de entrada
Z = z_func(X, Y)      # se eval√∫a la funci√≥n en la matriz de entrada

plt.figure(figsize=(5,5))              # tama√±o de imagen
plt.imshow(Z, cmap = 'inferno')        # grafico y paleta de colores
im_ratio = Z.shape[0]/A.shape[1]       # ancho y alto de la imagen
#plt.colorbar(fraction=0.046*im_ratio)  # barra de colores
plt.show()
```
