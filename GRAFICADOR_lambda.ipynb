<center> <h1> GRAFICADOR DE FUNCIONES </h1> </center> 

Este $script$ o $algoritmo$ hace las veces de graficador mediante el uso de funciones de tipo $lambda$, las cuales de definen mediante expresiones matemáticas en una sola línea.

Los parámetros a definir son:
1. La expresión matemática $f(x)$, las cuales deben seguir las reglas de sintaxis de Python.
2. Los argumentos $(x)$, los cuales son los parámetros o las variables que se reciber como entrada.

```python
# Librerías:
import numpy as np
import matplotlib.pyplot as plt

F1 = lambda x: pow(x,2) # función de tipo lambda

#F2 = lambda x: (np.cos(x))**3 # función de tipo lambda

x = np.arange(-5,5,0.5) # entradas de la función

# Se opera la función y se guardan las salidas en una variable:
FS1 = F1(x)

#FS2 = F2(x)

# Impresión de las salidas:
#print(FS1)
#print(' ')
#print(FS2)

# Gráfico de las salidas como puntos:
plt.plot(x, FS1, ".", markersize=6, color="black") 
#plt.plot(x, FS2, ".", markersize=5, color="red")  

# Gráfico de salidas como líneas:
plt.plot(x, FS1, ls="--", linewidth=1, color="black") 
#plt.plot(x, FS2, ls="--", linewidth=.5, color="black") 

# Graficar ejes:
#plt.hlines(y=0, xmin=-5, xmax=5)
#plt.vlines(x=0, ymin=-5, ymax=5)

plt.xlabel('Tiempo en s')   # Etiqueta eje horizontal
plt.ylabel('Temperatura en °C')    # Etiqueta eje vertical
plt.legend(['F1'])
#plt.grid()        # Área de trazado con cuadrícula
plt.savefig('sinwave.png')  
plt.show()        #  Gráfico como salida
```
