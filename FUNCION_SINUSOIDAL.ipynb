```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la función
amplitud = 1.0
frecuencia = 1.0  # en Hz
fase = 1/2*np.pi  # en radianes

t = np.linspace(0, 1/2*np.pi, 200)  # Serie de puntos en el tiempo

y = amplitud * np.sin((2*np.pi*frecuencia*t) + fase) #Función sinusoidal

# Graficar la función
#plt.plot(t, y)
plt.plot(t, y, ".", markersize=5, color="blue") 
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Función sinusoidal')
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100)  # Serie de puntos en el tiempo

y = amplitud * np.sin(t) #Función sinusoidal

# Graficar la función
#plt.plot(t, y)
plt.plot(t, y, ".", markersize=5, color="blue") 
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.title('Función sinusoidal')
plt.show()
```
